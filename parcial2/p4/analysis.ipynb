# Conclusiones

### Es evidente que la eficiencia de la recurion de cola y la traduccion iterativa funciona de una manera muchisimo mas optima que la funcion recursiva.

### Estos resultados se pueden explicar por la forma en que cada algoritmo aborda el problema. El algoritmo iterativo resuelve el problema de manera secuencial, un paso a la vez. El algoritmo de recursin resuelve el problema dividiendo el problema en subproblemas ms pequeos, y luego llamando recursivamente a s mismo para resolver esos subproblemas. El algoritmo de recursin de cola resuelve el problema de manera similar al algoritmo de recursin, pero mantiene una cola de subproblemas no resueltos.

### Para el caso de los N evaluados, estos son relativamente pequeos. Esto significa que el algoritmo de recursin puede resolver los subproblemas sin tener en un costo significativo. Sin embargo, a medida que N aumenta, el algoritmo de recursin comienza a incurrir en un costo significativo al crear y resolver subproblemas. El algoritmo iterativo, por otro lado, no tiene este costo, ya que resuelve el problema de manera secuencial.
