# Conclusiones

### Es evidente que la eficiencia de la recurion de cola y la traduccion iterativa funciona de una manera muchisimo mas optima que la funcion recursiva.

### Estos resultados se pueden explicar por la forma en que cada algoritmo aborda el problema. El algoritmo iterativo resuelve el problema de manera secuencial, un paso a la vez. El algoritmo de recursión resuelve el problema dividiendo el problema en subproblemas más pequeños, y luego llamando recursivamente a sí mismo para resolver esos subproblemas. El algoritmo de recursión de cola resuelve el problema de manera similar al algoritmo de recursión, pero mantiene una cola de subproblemas no resueltos.

### Para el caso de los N evaluados, estos son relativamente pequeños. Esto significa que el algoritmo de recursión puede resolver los subproblemas sin tener en un costo significativo. Sin embargo, a medida que N aumenta, el algoritmo de recursión comienza a incurrir en un costo significativo al crear y resolver subproblemas. El algoritmo iterativo, por otro lado, no tiene este costo, ya que resuelve el problema de manera secuencial.
